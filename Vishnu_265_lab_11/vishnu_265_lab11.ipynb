```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('pharmacy_management_data.csv')


print(df.info())
print(df.describe())


df.plot(x='Date', y='Quantity', kind='line')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.title('DATE VS QUANTITY')
plt.show()


df.plot(x='Medication ID', y='Quantity', kind='scatter')
plt.xlabel('Medication ID')
plt.ylabel('Quantity')
plt.title('Medication ID v Quantity')
plt.show()
```

```python
# Create a bar graph

plt.bar(df['Supplier Name'], df['Quantity'])
df=df.head(5)
plt.xlabel('Medication ID')
plt.ylabel('Quantity')
plt.title('Bar graph of product category vs sales')
plt.show()
```
