```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('pharmacy_management_data.csv')


print(df.info())
print(df.describe())


df.plot(x='Date', y='Quantity', kind='line')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.title('DATE VS QUANTITY')
plt.show()


df.plot(x='Medication ID', y='Total Amount', kind='scatter')
plt.xlabel('Medication ID')
plt.ylabel('Total Amount')
plt.title('Medication ID v Total Amount')
plt.show()
```
