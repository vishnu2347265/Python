```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
import contextily as ctx

#ShapeFile 

states_geo = gpd.read_file('India Shape\india_ds.shp')
states_geo.head(15)
```

```python
states_geo.plot()
```

```python
#read csv file
data = pd.read_csv('GOI.csv')
data.head(15)
```

```python
#let's plot the map with the literacy data

merged_data = states_geo.merge(data,left_on='STATE',right_on='Country/ States/ Union Territories Name',how='left')
merged_data.head()
```

```python

# Task 1: Explore literacy rates across Union Territories for 2001 and 2011
ut_data = merged_data[merged_data['Category'] == 'Union Territory']
plt.figure(figsize=(10, 6))
plt.bar(ut_data['Country/ States/ Union Territories Name'], ut_data['Literacy Rate (Persons) - Total - 2001'], label='2001')
plt.bar(ut_data['Country/ States/ Union Territories Name'], ut_data['Literacy Rate (Persons) - Total - 2011'], label='2011')
plt.title('Literacy Rates Across Union Territories (2001 vs 2011)')
plt.xlabel('Union Territories')
plt.ylabel('Literacy Rate (%)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Task 2: Investigate relationship between urban and rural literacy rates
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Literacy Rate (Persons) - Rural - 2011'], merged_data['Literacy Rate (Persons) - Urban - 2011'])
plt.title('Relationship between Urban and Rural Literacy Rates (2011)')
plt.xlabel('Rural Literacy Rate (%)')
plt.ylabel('Urban Literacy Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Task 3: Visualize data using appropriate graphs
plt.figure(figsize=(10, 6))
plt.hist(merged_data['Literacy Rate (Persons) - Total - 2011'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Literacy Rates in Indian States (2011)')
plt.xlabel('Total Literacy Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Task 4: Create an Indian state map to highlight level variations of Total Literacy of the year 2011
def categorize_level(literacy_rate):
    if literacy_rate <= 25:
        return 'Poor'
    elif 26 <= literacy_rate <= 50:
        return 'Fair'
    elif 51 <= literacy_rate <= 75:
        return 'Good'
    else:
        return 'Excellent'

merged_data['Literacy Level'] = merged_data['Literacy Rate (Persons) - Total - 2011'].apply(categorize_level)
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
merged_data.plot(column='Literacy Level', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
plt.title('Total Literacy Level in Indian States (2011)')
plt.show()
```
