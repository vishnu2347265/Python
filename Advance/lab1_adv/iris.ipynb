```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['species'] = iris.target_names[iris.target]
```

```python

# Bar Chart with 2 variables
plt.figure(figsize=(12, 6))
sns.barplot(x='species', y='sepal length (cm)', data=iris_df, ci=None)
plt.title('Bar Chart of Sepal Length for Iris Species')
plt.show()
```

```python

# Bar Chart with 3 variables
plt.figure(figsize=(14, 6))
sns.barplot(x='species', y='sepal length (cm)', hue='target', data=iris_df)
plt.title('Bar Chart of Sepal Length for Iris Species with Target'),   
plt.show()
```

```python

# Scatter Plot with 2 variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_df, hue='species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# Scatter Plot with 3 variables
plt.figure(figsize=(12, 8))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=iris_df)
plt.title('Scatter Plot of Sepal Length vs Sepal Width with Target')
plt.show()
```

```python
# Histogram with 2 variables
plt.figure(figsize=(10, 6))
sns.histplot(data=iris_df, x='sepal length (cm)', kde=True)
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# Histogram with 3 variables
plt.figure(figsize=(12, 8))
sns.histplot(data=iris_df, x='sepal length (cm)', hue='target', kde=True)
plt.title('Histogram of Sepal Length with Target')
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(8, 8))
iris_df['target'].value_counts().plot.pie(autopct='%1.1f%%', labels=iris.target_names)
plt.title('Pie Chart of Iris Species Distribution')
plt.show()
```

```python

# Pair plot
sns.pairplot(iris_df, hue='species', height=2.5)
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
# Bubble Plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', size='petal length (cm)', data=iris_df, hue='species')
plt.title('Bubble Plot of Sepal Length vs Sepal Width with Petal Length')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='species', y='sepal length (cm)', data=iris_df)
plt.title('Box Plot of Sepal Length for Iris Species')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal length (cm)', data=iris_df)
plt.title('Violin Plot of Sepal Length for Iris Species')
plt.show()
```

```python
# Area Plot
iris_df.drop('target', axis=1).plot.area(stacked=False, figsize=(12, 8))
plt.title('Area Plot of Iris Features')
plt.show()
```
