```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("D:\Progrmming\Python\Advance\Lab1\Q8_MortalityDataset.csv")

fig =  plt.figure()

ax = plt.axes(projection='3d')

alive = data[data['MORT'] == 'alive']
dead = data[data['MORT']=='dead']

ax.scatter(alive['HEIGHT'],alive['WEIGHT'],alive['AGE'],c="red",marker="o", label="Alive")
ax.scatter(dead['HEIGHT'],dead['WEIGHT'],dead['AGE'],c="black",marker="x", label="Dead")
ax.set_xlabel("Height")
ax.set_ylabel("WEIGHT")
ax.set_zlabel("AGE")
ax.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("D:\Progrmming\Python\Advance\Lab1\Q8_MortalityDataset.csv")

fig = plt.figure(figsize=(15,10))
ax = plt.axes(projection='3d')
a = data[data['BLOOD'] == 'a']
b = data[data['BLOOD'] == 'b']
o = data[data['BLOOD'] == 'o']
ab = data[data['BLOOD'] == 'ab']
ax.bar(a['AGE'],a['CHOL'],zdir='y',zs=0,color='pink',label='a')
ax.bar(b['AGE'],b['CHOL'],zdir='y',zs=1,color='red',label='b')
ax.bar(ab['AGE'],ab['CHOL'],zdir='y',zs=2,color='orange',label='ab')
ax.bar(o['AGE'],o['CHOL'],zdir='y',zs=3,color='purple',label='b')
ax.set_title('3D Bar plot')
ax.set_xlabel('AGE')
ax.set_ylabel('Blood')
ax.set_zlabel('CHOL')
ax.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(20, 18))
# Plot for alive individuals
ax1 = fig.add_subplot(221, projection='3d')
ax1.bar(a['HEIGHT'], a['CHOL'],zdir='y', zs=0, color='pink', label='BloodGroup-A')
ax1.set_xlabel('Height')
ax1.set_ylabel('BloodGroup-A')
ax1.set_zlabel('CHOL')
ax1.set_title('3D Bar Plot for BloodGroup-A')
ax1.legend()

# Plot for dead individuals
ax2 = fig.add_subplot(222, projection='3d')
ax2.bar(b['HEIGHT'], b['CHOL'],zdir='y', zs=1, color='red', label='BloodGroup-B')
ax2.set_xlabel('Height')
ax2.set_ylabel('BloodGroup-B')
ax2.set_zlabel('CHOL')
ax2.set_title('3D Bar Plot for BloodGroup-B')
ax2.legend()

# Plot for dead individuals
ax2 = fig.add_subplot(223, projection='3d')
ax2.bar(o['HEIGHT'], o['CHOL'],zdir='y', zs=2, color='orange', label='BloodGroup-B')
ax2.set_xlabel('Height')
ax2.set_ylabel('BloodGroup-O')
ax2.set_zlabel('CHOL')
ax2.set_title('3D Bar Plot for BloodGroup-O')
ax2.legend()

# Plot for dead individuals
ax2 = fig.add_subplot(224, projection='3d')
ax2.bar(ab['HEIGHT'], ab['CHOL'],zdir='y', zs=3, color='purple', label='BloodGroup-AB')
ax2.set_xlabel('Height')
ax2.set_ylabel('BloodGroup-AB')
ax2.set_zlabel('CHOL')
ax2.set_title('3D Bar Plot for BloodGroup-AB')
ax2.legend()

plt.show()
```
