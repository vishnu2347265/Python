```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
cap = cv.VideoCapture('video.mp4')
while(cap.isOpened()):
    ret,frame = cap.read()
    print(ret,frame)
    if ret == True:
        cv.imshow('Frame',frame)
        if cv.waitKey(25) & 0xFF == ord('q'):
            break
    else:
            break
cap.release()
cv.destroyAllWindows()
```

```python
import os
cam =  cv.VideoCapture('video.mp4')
try:
    if not os.path.exists('data'):
        os.makedirs('data')

except OSError:
    print("Error: creating directory of data")

currentframe = 0

while(True):

    ret,frame = cam.read()

    if ret:
        name = './data/frame' + str(currentframe) + '.jpg'
        print ('Creating...'+ name)
        cv.imwrite(name,frame)
        currentframe += 1
    else:
        break
cam.release()
cv.destroyAllWindows()
```
