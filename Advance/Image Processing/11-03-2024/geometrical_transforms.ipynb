Relative resizsing

```python
from PIL import Image,ImageOps
size = (100,150)
img =Image.open("picture.jpeg")
print("returns the tuple height and weight of the image",img.size)
ImageOps.contain(img,size).save("imageops_contain.png")
ImageOps.fit(img,size).save("imageops_fit.png")
ImageOps.pad(img,size,color="#f00").save("imageops_pad.png")
ImageOps.grayscale(img).save("imageops_gray.jpg")
```

```python
img.thumbnail(size)
img.save("imageops_thumnnail.png")
```
