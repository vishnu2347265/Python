```python
import cv2 as cv
image = cv.imread("flower.jpeg")
import matplotlib.pyplot as plt
import numpy as np
```

```python
brightness = 8
contrast = 1000000
img2 = cv.addWeighted(image,contrast,np.zeros(image.shape,image.dtype),0,brightness)


plt.subplot(121),plt.title('Orginl'),plt.imshow(image)
plt.subplot(122),plt.title('Brightness and contrast'),plt.imshow(img2)
plt.show()
```

```python
kernal = np.array([[0,-1,0],[-1,5,-1],[1,-1,0]])
kernal1 = np.ones((5,5),np.float32)/30
sharpened_image = cv.filter2D(src=image,ddepth=-1,kernel=kernal)
blur_image = cv.filter2D(image,-1,kernal1)
plt.subplot(221),plt.title('Orginl'),plt.imshow(image)
plt.subplot(222),plt.title('Sharpening'),plt.imshow(sharpened_image)
plt.subplot(223),plt.title('Blur'),plt.imshow(blur_image)
plt.show()
```

```python
image = cv.imread('picture.jpeg')
plt.figure(figsize=(10,10))
#remove noise
filtered_img1 = cv.medianBlur(image,11)
filtered_img2 = cv.GaussianBlur(image,(7,7),0)

#plot
plt.subplot(131),plt.title(f'Original Image'), plt.imshow(image)
plt.subplot(132),plt.title(f'median blur Image'), plt.imshow(filtered_img1)
plt.subplot(133),plt.title(f'gussianBlur Image'), plt.imshow(filtered_img2)
```

```python
Image=cv.cvtColor(image,cv.COLOR_RGB2HSV)#onvert the image from RGB to HSV color space.

Image[:,:,0]=image[:,:,0]*0.7 #Decrease the intensity of the red channel by 30%.
Image[:,:,1]=image[:,:,1]*0.7 # Decrease the intensity of the green channel by 30%.
Image[:,:,2]=image[:,:,2]*0.7 #Decrease the intensity of the blue channel by 30%.

image2=cv.cvtColor(Image,cv.COLOR_HSV2BGR)

plt.subplot(121),plt.title("Original"),plt.imshow(image)
plt.subplot(122),plt.title("Enhanced Color"),plt.imshow(image2)

```
