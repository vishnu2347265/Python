```python
#get supported image format
supported_formats = Image.registered_extensions()

#print the supported formats
print("Supported Formats:")
for format_name,extensions in supported_formats.items():
    print(f"{format_name}:{','.join(extensions)}")
```

```python
img = Image.open("child-3473596_640.jpg")
#crop subrectangle from image

box = (100,100,400,400)
#coordinates are (top,left,bottom,right)

portion= img.crop(box)
portion.show()
```

```python
#processing a subrectangle ,and pasting it back
portion = portion.transpose(Image.Transpose.FLIP_LEFT_RIGHT)
im.paste(portion,box)
# portion
im
```
