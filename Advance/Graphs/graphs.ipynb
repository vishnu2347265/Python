`````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('UScereal.csv')
`````

`````python
# Bar Charts using 2 variables
plt.figure(figsize=(10, 5))
sns.barplot(x='mfr', y='calories', data=df)
plt.title('Calories by Manufacturer')
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x='mfr', y='protein', data=df)
plt.title('Protein by Manufacturer')
plt.show()
`````
