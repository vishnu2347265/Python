```python
# using basemap package do the following
# locate karnataka and  plot bangalore in the karnataka map
# apply all the functions 

import numpy as np
from numpy import linspace
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from scipy import signal
from mpl_toolkits.basemap import Basemap
```

```python

plt.figure(figsize=(5,5))
m=Basemap(projection='merc',resolution='c',lat_0=15.3173,lon_0=75.7139)
m.bluemarble(scale=0.5)
plt.title("Coastlines",fontsize=20)
x,y=m(12.971599,77.594566)
plt.plot(x,y,'ok',markersize=3)
plt.text(x,y,'bangalore',fontsize=12)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

map = Basemap(projection='merc', lat_0=15, lon_0=75,
              resolution='l', area_thresh=1000.0,
              llcrnrlon=72.0, llcrnrlat=11.0,
              urcrnrlon=78.5, urcrnrlat=18.5)

map.drawcountries()
map.drawstates()
map.fillcontinents(color='lightgray', lake_color='aqua')
map.drawmapboundary(fill_color='aqua')

map.drawparallels(range(0, 40, 2), labels=[1, 0, 0, 0])
map.drawmeridians(range(60, 100, 2), labels=[0, 0, 0, 1])

bangalore_lon, bangalore_lat = 77.5946, 12.9716
x, y = map(bangalore_lon, bangalore_lat)
map.plot(x, y, 'ro', markersize=6)

plt.title('Karnataka Map with Bangalore')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap


map = Basemap(projection='lcc', width=1200000, height=900000,
              resolution='l', lat_0=15, lon_0=75,
              llcrnrlon=72.0, llcrnrlat=11.0,
              urcrnrlon=78.5, urcrnrlat=18.5)


map.drawcoastlines()
map.drawcountries()
map.drawstates()
map.drawrivers()
map.fillcontinents(color='coral',lake_color='aqua')
map.drawmapboundary(fill_color='aqua')

map.drawparallels(range(0, 40, 2), labels=[1, 0, 0, 0])
map.drawmeridians(range(60, 100, 2), labels=[0, 0, 0, 1])
bangalore_lon, bangalore_lat = 77.5946, 12.9716
x, y = map(bangalore_lon, bangalore_lat)
map.plot(x, y, 'ro', markersize=6)
plt.text(x,y,'bangalore',fontsize=12)
plt.title('Karnataka Map with Bangalore')
plt.show()
```
