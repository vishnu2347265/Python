```python
import numpy as np
from numpy import linspace
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D

data= pd.read_csv("D:\Progrmming\Python\Advance\Graphs\Iris.csv")


fig = plt.figure()
ax = plt.axes(projection='3d')
x = data['PetalWidthCm'].sort_values()
y = data['PetalLengthCm'].sort_values()

z = data['SepalLengthCm'].sort_values()
#plotting
ax.plot3D(x,y,z,c='black')
ax.set_title('3d scatter plot')
ax.set_xlabel('PetalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_zlabel('SepalLengthCm')
plt.show()
```

```python
#mpl.rcParams['font.size']=10
df= pd.read_csv("D:\Progrmming\Python\Advance\Graphs\Iris.csv")

setosa=df[df['Species']=='Iris-setosa']
versicolor=df[df['Species']=='Iris-versicolor']
virginica=df[df['Species']=='Iris-virginica']

fig = plt.figure
ax=fig.add_subplot(111,projection='3d')
ax.bar(setosa['Id'],setosa['PetalWidthCm'],zs=1,zdir='x',color="red",alpha=0.8)
ax.bar(versicolor['Id'],versicolor['PetalWidthCm'],zs=2,zdir='x',color="red",alpha=0.8)
ax.bar(virginica['Id'],virginica['PetalWidthCm'],zs=3,zdir='x',color="red",alpha=0.8)
ax.set_xlabel('Species')
ax.set_ylabel('Id')
ax.set_zlabel('PetalWidthCm')
plt.show()
```
