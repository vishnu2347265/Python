1. Create a form using Tkinter on your Domain details. (Domain name, Student name and other
details, with text boxes, radio buttons, menu and the required widgets you can use)

```python
import tkinter as tk
from tkinter import messagebox

# Create a new window with the title "Pharmacy Management System"
window = tk.Tk()
window.title("Pharmacy Management System")

# Create a function to submit the form data
def submit_form():
    customer_name = ent_customer_name.get()
    phone_number = ent_phone_number.get()
    medication_name = ent_medication_name.get()
    payment_method = payment_method_var.get()
    
    # Here, you can process and store the form data as needed.
    
    # Display the submitted data in a popup message box
    messagebox.showinfo("Form Submission", f"Customer Name: {customer_name}\nPhone Number: {phone_number}\nMedication Name: {medication_name}\nPayment Method: {payment_method}")

# Create a new frame to contain the widgets for entering customer information
frm_customer = tk.Frame(relief=tk.SUNKEN, borderwidth=3)
frm_customer.pack()

# Create the Label and Entry widgets for "Customer Name"
lbl_customer_name = tk.Label(master=frm_customer, text="Customer Name:")
lbl_customer_name.pack()
ent_customer_name = tk.Entry(master=frm_customer, width=30)
ent_customer_name.pack()

# Create the Label and Entry widgets for "Phone Number"
lbl_phone_number = tk.Label(master=frm_customer, text="Phone Number:")
lbl_phone_number.pack()
ent_phone_number = tk.Entry(master=frm_customer, width=30)
ent_phone_number.pack()

# Create the Label and Entry widgets for "Medication Name"
lbl_medication_name = tk.Label(master=frm_customer, text="Medication Name:")
lbl_medication_name.pack()
ent_medication_name = tk.Entry(master=frm_customer, width=30)
ent_medication_name.pack()

# Create the Label and Entry widgets for "Payment Method"
lbl_payment_method = tk.Label(master=frm_customer, text="Payment Method:")
lbl_payment_method.pack()

payment_method_var = tk.StringVar()
payment_method_var.set("Cash")  # Default selection

payment_methods = ["Cash", "Credit Card", "Online Payment"]

for method in payment_methods:
    rb_payment_method = tk.Radiobutton(
        master=frm_customer,
        text=method,
        variable=payment_method_var,
        value=method
    )
    rb_payment_method.pack()

# Create a Submit button to submit the form
btn_submit = tk.Button(
    master=frm_customer,
    text="Submit",
    command=submit_form
)
btn_submit.pack()

# Start the Tkinter main loop
window.mainloop()
```

2. Visualize your dataset using Matplotlib and write a program to create line graph, bar graph,
scatter plot and correlation graph or heat map.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Dataset.csv')
print(df)
```

```python
df = pd.read_csv('Dataset.csv')

df = df.tail(100)
plt.figure(figsize=(12, 6))
plt.plot(df['volume'], df['price'])
plt.xlabel('Date')
plt.ylabel('Amount Spent')
plt.title('Line Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(df['EAN_code'], df['num_review'], marker='o', linestyle='-', label='Number of Reviews')
plt.title('Number of Reviews Over EAN Codes')
plt.xlabel('EAN Code')
plt.ylabel('Number of Reviews')
plt.legend()
plt.grid()
plt.show()
```

```python
# Bar Graph (Example: Plotting 'overall_rate' by 'brand')
plt.figure(figsize=(18, 6))
sns.barplot(x='brand', y='overall_rate', data=df)
plt.title('Average Overall Rating by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Overall Rating')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter Plot (Example: Plotting 'product_qual' vs. 'customer_rate')
plt.figure(figsize=(8, 6))
plt.scatter(df['customer_rate'], df['product_qual'], color='green', marker='o')
plt.title('Scatter Plot: Quantity Purchased vs. Menu list price')
plt.xlabel('Quantity Purchased')
plt.ylabel('Menu_Item_Price')
plt.show()
```

```python

# Correlation Heatmap (Example: Correlation between numerical columns)
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
