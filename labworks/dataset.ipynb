```python
import pandas as pd
from faker import Faker
import random

# Create a Faker instance
fake = Faker()

# Create empty lists to store data for each column
medication_data = []
patient_data = []

# Generate fake data and populate the lists
for _ in range(100):  # You can adjust the number of rows as needed
    medication_data.append([
        fake.word(),
        fake.random_int(min=1000, max=9999),
        fake.random_element(elements=("5mg", "10mg", "20mg", "50mg")),
        fake.company(),
        fake.date_this_century(),
        round(random.uniform(1.0, 100.0), 2),  # Generate a random float
        fake.random_int(min=1, max=1000),
    ])
    
    patient_data.append([
        fake.name(),
        fake.random_int(min=10000, max=99999),
    ])

# Create DataFrames from the lists
medication_df = pd.DataFrame(medication_data, columns=[
    "Medication Name",
    "Medication ID",
    "Dosage Strength",
    "Manufacturer",
    "Expiry Date",
    "Unit Price",
    "Quantity in Stock",
])

patient_df = pd.DataFrame(patient_data, columns=[
    "Patient Name",
    "Patient ID",
])

# Concatenate the DataFrames
data = pd.concat([medication_df, patient_df], axis=1)

# Display the first few rows of the generated dataset
print(data.head(100))
```

```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

# Create an empty list to store data for each row
data = []

# Define the number of rows you want in your dataset
num_rows = 100  # Adjust this as needed

# Generate random data for each row and append it to the list
for _ in range(num_rows):
    medication_name = fake.word()
    medication_id = fake.random_int(min=1000, max=9999)
    quantity = fake.random_int(min=1, max=100)
    unit_price = round(random.uniform(5.0, 100.0), 2)
    total_amount = round(quantity * unit_price, 2)
    
    data.append({
        "Date": fake.date_between(start_date='-1y', end_date='today'),
        "Medication Name": medication_name,
        "Medication ID": medication_id,
        "Quantity": quantity,
        "Unit Price": unit_price,
        "Total Amount": total_amount,
        "Patient Name": fake.name(),
        "Prescribing Doctor": fake.name(),
        "Payment Method": fake.random_element(elements=("Cash", "Credit Card", "Debit Card", "Insurance")),
        "Payment Status": fake.random_element(elements=("Paid", "Pending")),
        "Supplier Name": fake.company(),
        "Tags/Labels": fake.word(),
        "Notes": fake.sentence()
    })

# Create a DataFrame from the list of dictionaries
data_df = pd.DataFrame(data)

# Display the first few rows of the generated dataset
print(data_df.head())

# Save the dataset to a CSV file
data_df.to_csv("pharmacy_management_data.csv", index=False)
```
